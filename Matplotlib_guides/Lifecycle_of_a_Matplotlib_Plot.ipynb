Start with a figure (`fig`) and axis (`ax`).  The "figure" is the entire canvas on which the graph(s) appear.  The "axis" is a particular graph; it does not mean the x-axis or y-axis, but all components.  We will use the `subplots` command even when we have only one subplot.  Here `(1,1)` refers to one row and one column.

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
```

We don't need to give the keywords `nrows` and `ncols`:

```python
fig, ax = plt.subplots(1, 1)
```

This is the default aspect ratio.  What if we want to specify the size?

```python
fig, ax = plt.subplots(1, 1, figsize=(5,5))
```

What about having two graphs?  We call them subplots.

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))
```

Ok, back to one graph.  Now lets plot some lines:

```python
t = np.linspace(-2 * np.pi, 2 * np.pi, 1000) # 1000 pts from -2\pi to 2\pi
x = np.linspace(0, 14, 100)
fig, ax = plt.subplots(1, 1, figsize=(5,5))
for i in range(1, 7):
    ax.plot(x, np.sin(x + i * .5) * (7 - i))
```

Now add some axis labels and turn on legends and make it a function:

```python
def make_my_plot(ax=None):
    if ax is None:
        ax = plt.gca()

    x = np.linspace(0, 14, 100)
    #fig, ax = plt.subplots(1, 1, figsize=(5,5))
    for i in range(1, 7):
        ax.plot(x, np.sin(x + i * .5) * (7 - i),label=i)
    ax.set_xlabel('$x$')   
    ax.set_ylabel('$f(x)$')
    ax.legend()   # or 'upper right', 'lower left', etc. --> shift-tab
    ax.set_xlim(-1,15)
    ax.set_ylim(-7,7)
    return ax
```

What styles could we use?

```python
sorted(mpl.style.available)
```

```python
mpl.style.use('classic')
make_my_plot()
```

```python
mpl.style.use('fivethirtyeight')
make_my_plot()
```

```python
mpl.style.use('seaborn')
make_my_plot()
```

```python
mpl.style.use('bmh')
make_my_plot()
```

```python
# this doesn't actually work the way we want
#  Probably can't have different styles in subplots
from math import ceil
num_styles = len(sorted(mpl.style.available))
style_cols = 4
style_rows = int(ceil(num_styles/4))
fig, axes = plt.subplots(style_rows, style_cols, figsize=(15,3*style_rows))
for i,my_style in enumerate(sorted(mpl.style.available)):
    mpl.style.use(my_style)
    make_my_plot(axes[i % style_rows][i % style_cols])
fig.tight_layout()
```
