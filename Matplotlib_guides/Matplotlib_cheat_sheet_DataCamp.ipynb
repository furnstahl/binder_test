## Plot anatomy and workflow

We'll describe the parts of a generic plot as we generate some.  The steps in the workflow are:

0. Import plotting module(s)
1. Prepare data.
2. Create plot.
3. Plot.
4. Customize plot.
5. Save plot.
6. Show plot.

Let's first step through them with a simple example and then return to each in turn.

```python
%matplotlib inline
# 0. Import plotting modules
import matplotlib.pyplot as plt
# 1. Prepare data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
# 2. Create plot
fig = plt.figure()
# 3. Plot
ax = fig.add_subplot(111)
# 3,4. Customize plot
ax.plot(x, y, color='red', linewidth=3)
ax.scatter([2,4,6],
           [5,15,25],
           color='darkgreen',
           marker='^',
          )
ax.set_xlim(1, 6.5)
# 5. Save plot
plt.savefig('foo.png')
# 6. Show plot
plt.show()
```

## 2. Create plot

First the figure.

```python
import matplotlib.pyplot as plt
# Create a figure
#  fig, fig2, and fig3 refer to those particular figures.
fig = plt.figure()
fig2 = plt.figure(figsize=plt.figaspect(2.0))
fig3 = plt.figure(figsize=(5,5))  # 5 inches by 5 inches
```

In Matplotlib, Axes are the plots themselves.  A subplot is an axes on a grid; almost always that should be what you create, even if you only plan a single panel.

```python
fig = plt.figure()
fig.add_axes()
ax1a = fig.add_subplot(221)  # row-col-which one (so 2x2 and this is the first one)
ax3a = fig.add_subplot(212)  # we can use whatever name we want for the axes
ax1a.plot(x,y)
ax3a.plot(y,x)
```

```python
fig.add_axes()
ax1 = fig.add_subplot(221)  # row-col-which one (so 2x2 and this is the first one)
ax3 = fig.add_subplot(212)  # we can use whatever name we want for the axes
fig3, axes = plt.subplots(nrows=2,ncols=2)  # here axes is a 2x2 array of four axes in a grid 
fig4, axes2 = plt.subplots(ncols=3)  # here nrows is implicitly 1
```

```python
print(fig3.get_size_inches())  # dimensions of the figure in inches [horizontal vertical]
print(fig3.get_size_inches()*fig3.dpi)  # dimensions of the figure in pixels
```

## 3. Plotting routines

###  1d data

```python
fig, ax = plt.subplots()
lines = ax.plot(x,y,'b-',lw=2)  # Draw points with lines or markers connecting them
ax.scatter(x,y,color='red')       # Draw unconnected points, scaled or colored
```

```python
fig3, axes = plt.subplots(nrows=2,ncols=2)  # here axes is a 2x2 array of four axes in a grid 
axes[0,0].bar([1,2,3],[3,4,5]) # Plot vertical rectangles (constant width)
axes[1,0].barh([0.5,1,2.5],[0,1,2])  # Plot horizontal retangles (constant height)
axes[1,1].axhline(0.45)  # Draw a horizontal line across axes
axes[0,1].axvline(0.65)  # Draw a vertical line across axes
fig3.tight_layout()  # Try it with this commented out to see what it does
```

### Text & Annotations

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.text(2, 0.5,
        'Example Graph',
        style='italic')
ax.annotate("Sine",
           xy=(8,0),
           xycoords='data',
           xytext=(9.5,0.6),
           textcoords='data',
           arrowprops=dict(arrowstyle="->",
                          connectionstyle="arc3"),)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,y)
ax.text(2, 0.5,
        'Example Graph',
        style='italic')
ax.annotate("Sine",
           xy=(8,0),
           xycoords='data',
           xytext=(9.0,0.6),
           textcoords='data',
           fontsize=16,
           arrowprops=dict(width=3, headwidth=10, headlength=10, 
                          connectionstyle="arc3"),)
```

### Mathtext

### Units, Legends & Layouts

```python
fig, ax = plt.subplots()
ax.plot(x,y,lw=4)
ax.margins(x=0.5,y=2)  # add padding to a plot
ax.axis('equal')  # set aspect ratio to 1  (Doesn't seem to work here)
ax.set(xlim=(0,10.5), ylim=(-1.5,1.5))
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, lw=4, label='sine wave')
ax.set(xlim=(0,10.5), ylim=(-1.5,1.5))
ax.set(title='An Example Axes',
       ylabel='Y-Axis',
       xlabel='X-axis')
ax.legend(loc='best')
ax.xaxis.set(ticks=range(1,10),
             ticklabels=[3,100,-12,"foo","a","b","c"])
ax
```

```python
fig3, axes = plt.subplots(nrows=2,ncols=2)  # here axes is a 2x2 array of four axes in a grid 
axes[0,0].bar([1,2,3],[3,4,5]) # Plot vertical rectangles (constant width)
axes[1,0].barh([0.5,1,2.5],[0,1,2])  # Plot horizontal retangles (constant height)
axes[1,1].axhline(0.45)  # Draw a horizontal line across axes
axes[0,1].axvline(0.65)  # Draw a vertical line across axes
fig3.subplots_adjust(wspace=1.5,  # change these to see the effects
                     hspace=0.5,
                     left=0.125,
                     right=0.9,
                     bottom=0.1)
fig.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, lw=4, label='sine wave')
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_position(('outward',20))
```
